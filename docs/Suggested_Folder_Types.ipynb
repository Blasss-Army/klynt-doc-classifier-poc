# M&A Folder Structure (Suggested Data Room Categories)

- **00_Admin_&_Process**
  - Timeline, process letter, contacts, trackers

- **01_Confidentiality**
  - NDA, permitted recipients list, clean team (if applicable)

- **02_Marketing_Materials**
  - Teaser, CIM/IM, marketing presentations

- **03_Indicative_Offers**
  - EOIs, LOIs/Term_Sheets, bid letters and versions

- **04_Exclusivity**
  - Exclusivity Agreement and annexes

- **05_Due_Diligence**
  - **05.1_Financial_DD**
  - **05.2_Tax_DD**
  - **05.3_Legal_DD**
  - **05.4_Commercial_DD**
  - **05.5_IT_&_Cyber_DD**
  - **05.6_HR_DD**
  - **05.7_Environmental_DD**
  - **05.8_Quality_of_Earnings_(QoE)**
  - **05.9_Q&A_Log**
  - **05.10_Management_Presentations**

- **06_Definitive_Agreements**
  - SPA/APA, drafts, markups, signature versions

- **07_Disclosures**
  - Disclosure Letter, Disclosure Schedules

- **08_Insurance**
  - W&I Insurance, broker notes, policies

- **09_Closing**
  - Closing checklist, deliverables, closing sets

- **10_Transition_&_Post-Closing**
  - TSA, integration plan, post-close communications

- **11_Employment_&_Incentives**
  - Management/retention agreements, earn-outs, non-compete / non-solicit

- **12_Financing**
  - Commitment letters, financing term sheets, facility/loan agreements

- **13_Security_&_Intercreditor**
  - Security documents, guarantees, intercreditor agreement

- **14_Regulatory_&_Compliance**
  - Antitrust/competition filings, sector approvals, KYC/AML, board/shareholder resolutions
