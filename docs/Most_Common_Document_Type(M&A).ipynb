# Most Common Documents in M&A (Mergers & Acquisitions)

## 1) Preparation and marketing (sell-side / buyer outreach)

- **Teaser**: 1â€“2 pages, high-level overview, typically without clearly identifying the company.
- **CIM / Confidential Information Memorandum (IM)**: comprehensive business deck (market, financials, operations, risks, etc.).
- **Process letter / Process instructions**: timeline, bid submission guidelines, requirements, contact details, etc.
- **NDA / Non-Disclosure Agreement**: required to access confidential information.
- **Data Room Index**: folder structure/listing of documents in the data room.

## 2) Offers and initial negotiation

- **EOI / Indication of Interest**: preliminary, non-binding expression of interest with valuation range and key assumptions.
- **LOI / Letter of Intent** or **Term Sheet**: key commercial terms (price, structure, exclusivity, conditions, timeline). Typically **mostly non-binding**, except for certain clauses.
- **Exclusivity Agreement**: exclusivity arrangement (sometimes included within the LOI).

## 3) Due diligence

- **Vendor Due Diligence (VDD)** (if prepared by the seller): reports such as:
  - **Financial DD**
  - **Tax DD**
  - **Legal DD**
  - **Commercial DD**
  - (sometimes) **IT / Cyber / HR / Environmental DD**
- **Q&A log**: record of buyer questions and seller responses.
- **Management presentations**: presentations by the management team.
- **Quality of Earnings (QoE)** (very common in private equity): analysis of normalized EBITDA, adjustments, one-offs, etc.

## 4) Definitive agreements (signing and closing)

- **SPA / Share Purchase Agreement** (share deal) or **APA / Asset Purchase Agreement** (asset deal): main purchase agreement.
- **Disclosure Letter / Disclosure Schedules**: disclosures qualifying warranties/representations.
- **W&I Insurance** (if applicable): warranties & indemnities insurance policy.
- **Escrow Agreement**: holdback/deposit to cover potential claims.
- **Transition Services Agreement (TSA)**: post-close transitional services (IT, accounting, payroll, etc.).
- **Employment / Management agreements**: retention, incentives, earn-outs for key executives.
- **Non-compete / Non-solicit**: non-compete / non-solicitation undertakings.
- **Closing checklist**: list of conditions and deliverables required to close.

## 5) Financing and structure (if debt or PE-backed)

- **Commitment letter** and **financing term sheet**
- **Facility agreement / Loan agreement**
- **Security documents** (pledges, guarantees, etc.)
- **Intercreditor agreement** (if multiple layers of debt)

## 6) Approvals and compliance

- **Regulatory filings** (competition/antitrust, sector-specific approvals)
- **Board / shareholder resolutions**: approvals from the board and/or shareholders.
- **KYC / AML** (especially when banks or complex structures are involved)
