```python
import pandas as pd

def extract_xlsx_text(path: str) -> str:
    sheets = pd.read_excel(path, sheet_name=None)
    parts = []
    for name, df in sheets.items():
        parts.append(f"=== {name} ===")
        parts.append(df.astype(str).fillna("").to_string(index=False))
    return "\n".join(parts)

text = extract_xlsx_text("C:/Users/mamen/Desktop/Project/Assigment 3 - Klynt AG/data/messy_folder/payslip_oct_2025.xlsx")
print(text[:1000])
```
